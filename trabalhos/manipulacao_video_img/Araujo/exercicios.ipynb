```python
def plot_duas_im(im1, tit1, im2, tit2, tam):
    figure(figsize=tam)
    plt.subplot(1, 2, 1), plt.imshow(im1), plt.title(tit1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(1, 2, 2), plt.imshow(im2), plt.title(tit2)
    plt.xticks([]), plt.yticks([])
    plt.show()

def plot_tres_im(im1, tit1, im2, tit2, im3, tit3, tam):
    figure(figsize=tam)
    plt.subplot(1, 3, 1), plt.imshow(im1), plt.title(tit1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(1, 3, 2), plt.imshow(im2), plt.title(tit2)
    plt.xticks([]), plt.yticks([])
    plt.subplot(1, 3, 3), plt.imshow(im3), plt.title(tit3)
    plt.xticks([]), plt.yticks([])
    plt.show()
```

## função passada em sala para visualização de imagens

```python
plot_duas_im(im_dia1, "img dia 1", im_dia1, "img dia 1", (15,12))
```

## criei uma função para mostrar histogramas de 3 imagens

```python
def tres_im_histograma(im, im2, im3):
    im1_ = array(im)
    im2_ = array(im2)
    im3_ = array(im3)
    plt.hist(im1_.ravel())
    plt.hist(im2_.ravel())
    plt.hist(im3_.ravel())
```

## agora com função

```python
def zoom(im, linha, linha2, coluna, coluna2):
    imagem = im[linha:linha2, coluna:coluna2]
    plt.imshow(imagem)
```
