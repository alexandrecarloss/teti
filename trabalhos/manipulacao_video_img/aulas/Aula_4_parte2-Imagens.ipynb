```python
def plot_2img(img1, tit1, im2, tit2, tam):
    figure(figsize = tam)
    plt.subplot(1,2,1), plt.imshow(im1), plt.title(tit1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(1,2,2), plt.imshow(im2), plt.title(tit2)
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
def plot_3img(im1, tit1, im2, tit2, im3, tit3, tam):
    figure(figsize = tam)
    plt.subplot(1,3,1), plt.imshow(im1, cmap = "gray"), plt.title(tit1)
    plt.xticks([]), plt.yticks([])
    plt.subplot(1,3,2), plt.imshow(im2, cmap = "gray"), plt.title(tit2)
    plt.xticks([]), plt.yticks([])
    plt.subplot(1,3,3), plt.imshow(im3, cmap = "gray"), plt.title(tit3)
    plt.xticks([]), plt.yticks([])
    plt.show()
```

```python
im2 = array(Image.open(path_img + 'balao_montanhas_geladas.jpg'))
```

```python
plot_2img(im1, 'Imagem1', im2, 'Imagem2', (12, 10))
```

```python
im3 = array(Image.open(path_img + 'baleias_mar_gelo.jpg'))
```

```python
plot_3img(im1, 'Imagem1', im2, 'Imagem2', im3, 'Imagem3', (12, 10))
```
