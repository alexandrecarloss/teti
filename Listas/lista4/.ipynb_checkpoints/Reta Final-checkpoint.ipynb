## bibliotecas neccessárias

```python
import tkinter as tk
from tkinter import ttk
from tkinter import filedialog
import cv2
from PIL import Image, ImageTk
import os
import numpy as np
```

## Definição de kernels

```python
blur_kernel = np.ones((5, 5), np.float32) / 25
sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]], dtype=np.float32)
emboss_kernel = np.array([[-2, -1, 0],
                          [-1, 1, 1],
                          [0, 1, 2]], dtype=np.float32)
laplacian_kernel = np.array([[0, 1, 0],
                             [1, -4, 1],
                             [0, 1, 0]], dtype=np.float32)
sobel_kernel_x = np.array([[-1, 0, 1],
                           [-2, 0, 2],
                           [-1, 0, 1]], dtype=np.float32)
sobel_kernel_y = np.array([[-1, -2, -1],
                           [0, 0, 0],
                           [1, 2, 1]], dtype=np.float32)
```

## definição de funções

## ZOOM

```python
zoom_factor = 1.0
imagem_carregada = None
def zoom_in(event=None):
    global zoom_factor
    zoom_factor *= 1.1
    aplicar_zoom()
def zoom_out(event=None):
    global zoom_factor
    zoom_factor /= 1.1
    aplicar_zoom()
def aplicar_zoom():
    global imagem_carregada, canvas, img_container
    if imagem_carregada is not None:
        width, height = imagem_carregada.size
        new_width, new_height = int(width * zoom_factor), int(height * zoom_factor)
        img_zoomed = imagem_carregada.resize((new_width, new_height), Image.LANCZOS)
        imgtk_zoomed = ImageTk.PhotoImage(img_zoomed)
        canvas.itemconfig(img_container, image=imgtk_zoomed)
        canvas.imgtk = imgtk_zoomed
```

## escolha de filtro

```python
def aplicar_filtro(kernel):
    global imagem_carregada, canvas, img_container
    if imagem_carregada is not None:
        img = cv2.filter2D(np.array(imagem_carregada), -1, kernel)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        img = Image.fromarray(img)
        imgtk = ImageTk.PhotoImage(image=img)
        canvas.itemconfig(img_container, image=imgtk)
        canvas.imgtk = imgtk 
```

## escolha de evento a partir de teclas

```python
def keypress(event):
    key = event.char
    if key == '1':
        aplicar_filtro(blur_kernel)
    elif key == '2':
        aplicar_filtro(sharpen_kernel)
    elif key == '3':
        aplicar_filtro(emboss_kernel)
    elif key == '4':
        aplicar_filtro(laplacian_kernel)
    elif key == '5':
        aplicar_filtro(sobel_kernel_x)
    elif key == '6':
        aplicar_filtro(sobel_kernel_y)
    elif key == '7':
        imagem = cv2.Canny(np.array(imagem_carregada), 100, 200)
        imagem = cv2.cvtColor(imagem, cv2.COLOR_GRAY2RGB)  # Converter para RGB para exibir corretamente
        imagem = Image.fromarray(imagem)
        imgtk = ImageTk.PhotoImage(image=imagem)
        canvas.itemconfig(img_container, image=imgtk)
        canvas.imgtk = imgtk
```

```python
def mostrar_imagem_video(file_path, modo, root, tipo_arquivo):
    global imagem_carregada, imagem_original, canvas, img_container, blur_var, sharpen_var, emboss_var, laplacian_var, canny_var, sobel_var, cores_var, cinza_var, binario_var, cap, filtro_var, tons_var
    cap = cv2.VideoCapture(file_path)
    
    if not cap.isOpened():
        print("Erro: não foi possível abrir o arquivo de vídeo.")
        return
    
    # Configurações da segunda janela
    segunda_janela = tk.Toplevel(root)
    segunda_janela.title("Manipulação de Imagens e Vídeos")
    segunda_janela.attributes('-fullscreen', True)
    
    # PanedWindow para o painel superior
    paned_window_top = tk.PanedWindow(segunda_janela, orient=tk.HORIZONTAL, bd=4, relief='raised', bg='#e6f2ff')
    paned_window_top.pack(side=tk.TOP, fill=tk.X, pady=20)
    
    # Frame para a label e checkboxes
    frame_top = tk.Frame(paned_window_top, bg="#e6f2ff")
    paned_window_top.add(frame_top)
    
    # Label "Aplicar Filtro"
    label_filtro = tk.Label(frame_top, text="Aplicar Filtro", font=("Arial", 24), bg="#e6f2ff")
    label_filtro.pack(side=tk.LEFT, padx=20)
    
    # Variáveis para armazenar os estados das Radiobuttons
    filtro_var = tk.StringVar(value="nenhum")
    tons_var = tk.StringVar(value="nenhum")
    
    # Radiobuttons de filtros
    rbtn_blur = tk.Radiobutton(frame_top, text="Blur/Tecla1", variable=filtro_var, value="blur", bg="#e6f2ff", font=("Arial", 14))
    rbtn_blur.pack(side=tk.LEFT, padx=10)
    rbtn_sharpen = tk.Radiobutton(frame_top, text="Sharpen/Tecla2", variable=filtro_var, value="sharpen", bg="#e6f2ff", font=("Arial", 14))
    rbtn_sharpen.pack(side=tk.LEFT, padx=10)
    rbtn_emboss = tk.Radiobutton(frame_top, text="Emboss/Tecla3", variable=filtro_var, value="emboss", bg="#e6f2ff", font=("Arial", 14))
    rbtn_emboss.pack(side=tk.LEFT, padx=10)
    rbtn_laplacian = tk.Radiobutton(frame_top, text="Laplacian/Tecla4", variable=filtro_var, value="laplacian", bg="#e6f2ff", font=("Arial", 14))
    rbtn_laplacian.pack(side=tk.LEFT, padx=10)
    rbtn_canny = tk.Radiobutton(frame_top, text="Canny/Tecla5", variable=filtro_var, value="canny", bg="#e6f2ff", font=("Arial", 14))
    rbtn_canny.pack(side=tk.LEFT, padx=10)
    rbtn_sobel = tk.Radiobutton(frame_top, text="Sobel/Tecla7-8", variable=filtro_var, value="sobel", bg="#e6f2ff", font=("Arial", 14))
    rbtn_sobel.pack(side=tk.LEFT, padx=10)
    
    # Espaço entre radiobuttons e nova label
    espaco = tk.Label(frame_top, text="   ", bg="#e6f2ff")
    espaco.pack(side=tk.LEFT, padx=10)
    
    # Label "Tons de Imagem"
    label_tons_imagem = tk.Label(frame_top, text="Tons de Imagem", font=("Arial", 24), bg="#e6f2ff")
    label_tons_imagem.pack(side=tk.LEFT, padx=20)
    
    # Radiobuttons de tons de imagem
    rbtn_cores = tk.Radiobutton(frame_top, text="Cores", variable=tons_var, value="cores", bg="#e6f2ff", font=("Arial", 14))
    rbtn_cores.pack(side=tk.LEFT, padx=10)
    rbtn_cinza = tk.Radiobutton(frame_top, text="Cinza", variable=tons_var, value="cinza", bg="#e6f2ff", font=("Arial", 14))
    rbtn_cinza.pack(side=tk.LEFT, padx=10)
    rbtn_binario = tk.Radiobutton(frame_top, text="Binário", variable=tons_var, value="binario", bg="#e6f2ff", font=("Arial", 14))
    rbtn_binario.pack(side=tk.LEFT, padx=10)
    
    # Espaço entre radiobuttons e nova label
    espaco = tk.Label(frame_top, text="   ", bg="#e6f2ff")
    espaco.pack(side=tk.LEFT, padx=10)
    
    # Label "Vídeo"
    label_video = tk.Label(frame_top, text="Vídeo", font=("Arial", 24), bg="#e6f2ff")
    label_video.pack(side=tk.LEFT, padx=10)
    
    # Ícones de controle de vídeo
    btn_desacelerar = tk.Button(frame_top, text="Desacelerar", font=("Arial", 14), bg="#007acc", fg="#ffffff", command=lambda: print("Desacelerar"))
    btn_desacelerar.pack(side=tk.LEFT, padx=10)
    btn_pausar = tk.Button(frame_top, text="Pausar", font=("Arial", 14), bg="#007acc", fg="#ffffff", command=lambda: print("Pausar"))
    btn_pausar.pack(side=tk.LEFT, padx=10)
    btn_iniciar = tk.Button(frame_top, text="Iniciar", font=("Arial", 14), bg="#007acc", fg="#ffffff", command=lambda: print("Iniciar"))
    btn_iniciar.pack(side=tk.LEFT, padx=10)
    btn_acelerar = tk.Button(frame_top, text="Acelerar", font=("Arial", 14), bg="#007acc", fg="#ffffff", command=lambda: print("Acelerar"))
    btn_acelerar.pack(side=tk.LEFT, padx=10)
    btn_inverter = tk.Button(frame_top, text="Inverter", font=("Arial", 14), bg="#007acc", fg="#ffffff", command=lambda: print("Inverter"))
    btn_inverter.pack(side=tk.LEFT, padx=10)

    #Função para habilitar ou desabilitar botões de controle de vídeo
    def set_video_controls_state(state):
        
        btn_desacelerar.config(state=state)
        btn_pausar.config(state=state)
        btn_iniciar.config(state=state)
        btn_acelerar.config(state=state)
        btn_inverter.config(state=state)

    # Adicionar botões ao frame de vídeo
    """btn_desacelerar.pack(side=tk.LEFT, padx=34)
    btn_pausar.pack(side=tk.LEFT, padx=34)
    btn_iniciar.pack(side=tk.LEFT, padx=34)
    btn_acelerar.pack(side=tk.LEFT, padx=34)
    btn_inverter.pack(side=tk.LEFT, padx=34) """

    # Desativar ou ativar botões de controle de vídeo baseado no tipo de arquivo
    if tipo_arquivo == "Imagem":
        set_video_controls_state("disabled")
    elif tipo_arquivo == "Vídeo":
        set_video_controls_state("normal")

    # Canvas para exibir a imagem
    canvas = tk.Canvas(segunda_janela, bg="white")
    canvas.pack(fill=tk.BOTH, expand=True)

    def mostra_quadro():
        if not cap.isOpened():
            print("Erro ao capturar o quadro.")
            return

        ret, frame = cap.read()
        if ret:
            frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
            img = Image.fromarray(frame)
            # Obter a largura e altura da segunda janela
            screen_width = segunda_janela.winfo_screenwidth()
            screen_height = segunda_janela.winfo_screenheight()
            # Calcular o tamanho da imagem
            img_width = int(screen_width * 0.3)
            img_height = int(screen_height * 0.3)
            # Redimensionar a imagem
            img = img.resize((img_width, img_height), Image.LANCZOS)
            imgtk = ImageTk.PhotoImage(image=img)
            # Centralizar a imagem no canvas, com ajuste no eixo y
            x_centro = (screen_width - img_width) // 2
            y_centro = (screen_height - img_height) // 4  # Ajuste no eixo y para subir a imagem
            # Exibir a imagem no canvas
            img_container = canvas.create_image(x_centro, y_centro, anchor=tk.NW, image=imgtk)
            canvas.imgtk = imgtk  # Manter a referência para evitar a coleta de lixo

            # Continuar a reprodução
            if segunda_janela.winfo_exists():
                canvas.after(10, mostra_quadro)
            
        else:
            #print("Erro ao capturar o quadro!.")
            #return
            # Se chegou ao fim do vídeo, reiniciar
            cap.set(cv2.CAP_PROP_POS_FRAMES, 0)
            if segunda_janela.winfo_exists():
                canvas.after(10, mostra_quadro)
            else:
                cap.release()
        """if segunda_janela.winfo_exists():
            canvas.after(10, mostra_quadro)
        else:
            cap.release() """

    mostra_quadro()
    
    # PanedWindow para os botões na parte inferior
    paned_window_bottom = tk.PanedWindow(segunda_janela, orient=tk.HORIZONTAL, bd=4, relief='raised', bg='#e6f2ff')
    paned_window_bottom.pack(side=tk.BOTTOM, fill=tk.X, pady=20)

    # Frame para os botões
    frame_botoes = tk.Frame(paned_window_bottom, bg="#e6f2ff")
    paned_window_bottom.add(frame_botoes)

    # Função para posicionar botões proporcionalmente
    def pack_proportionally(widget, side, padx=10):
        widget.pack(side=side, padx=padx, expand=True)
    
    # Botão "Voltar"
    btn_voltar = tk.Button(frame_botoes, text="Voltar", font=("Arial", 14), bg="#007acc", fg="#ffffff", padx=30, pady=20, command=segunda_janela.destroy)
    pack_proportionally(btn_voltar, tk.LEFT)

    # Botão "Zoom+"
    btn_zoom_mais = tk.Button(frame_botoes, text="Zoom+", font=("Arial", 14), bg="#007acc", fg="#ffffff", padx=30, pady=20, command=lambda: print("Zoom In"))
    pack_proportionally(btn_zoom_mais, tk.LEFT)

    # Botão "Zoom-"
    btn_zoom_menos = tk.Button(frame_botoes, text="Zoom-", font=("Arial", 14), bg="#007acc", fg="#ffffff", padx=30, pady=20, command=lambda: print("Zoom Out"))
    pack_proportionally(btn_zoom_menos, tk.LEFT)

    # Botão "Recorte"
    btn_recorte = tk.Button(frame_botoes, text="Recorte", font=("Arial", 14), bg="#007acc", fg="#ffffff", padx=30, pady=20, command=lambda: print("Recortar"))
    pack_proportionally(btn_recorte, tk.LEFT)

    # Botão "Desfazer Recorte"
    btn_desfazer_recorte = tk.Button(frame_botoes, text="Desfazer Recorte", font=("Arial", 14), bg="#007acc", fg="#ffffff", padx=30, pady=20, command=lambda: print("Desfazer Recorte"))
    pack_proportionally(btn_desfazer_recorte, tk.LEFT)

    # Botão "Salvar"
    btn_salvar = tk.Button(frame_botoes, text="Salvar", font=("Arial", 14), bg="#007acc", fg="#ffffff", padx=30, pady=20, command=lambda: print("Salvar"))
    pack_proportionally(btn_salvar, tk.LEFT)

    # Botão "Sair"
    btn_sair = tk.Button(frame_botoes, text="Sair", font=("Arial", 14), bg="#007acc", fg="#ffffff", padx=30, pady=20, command=segunda_janela.destroy)
    pack_proportionally(btn_sair, tk.RIGHT)
```

```python
def selecionar_arq(modo, root):
    inicial_dir = os.path.expanduser("/home/wallace/Documentos/TETI/teti_imagens/")  # Define o diretório inicial como o diretório principal do usuário
    file_path = filedialog.askopenfilename(initialdir=inicial_dir, filetypes=[("All Files", "*.*"), ("Image files", "*.jpg *.png"), ("Video files", "*.mp4 *.avi")])
    if file_path:
        tipo_arquivo = "Imagem" if file_path.lower().endswith(('.png', '.jpg', '.jpeg')) else "Vídeo"
        mostrar_imagem_video(file_path, modo, root, tipo_arquivo)  
```

```python
def main():
    root = tk.Tk()
    root.title("Manipulação de Imagens e Vídeos")
    root.geometry("1200x900")
    
    # Estilo da janela principal
    root.configure(background="#e6f2ff")

    # Estilo dos componentes
    style = ttk.Style()
    style.configure("TLabel", background="#e6f2ff", foreground="#333333", padding=(15, 10, 15, 10), font=("Arial", 14))
    style.configure("TCombobox", foreground="#333333", padding=(15, 10, 20, 10), font=("Arial", 14))

    # Dropdown para selecionar o tipo de arquivo
    tipo_arquivo_label = ttk.Label(root, text="Selecione o Tipo de Arquivo", style="TLabel")
    tipo_arquivo_label.place(x=400, y=100)
    
    tipo_arquivo = ttk.Combobox(root, values=["Imagem", "Vídeo"], style="TCombobox")
    tipo_arquivo.place(x=600, y=100)

    # Dropdown para selecionar o modo
    modo_label = ttk.Label(root, text="Selecione o Modo", style="TLabel")
    modo_label.place(x=400, y=200)
    
    modo = ttk.Combobox(root, values=["Independente", "Cascata"], style="TCombobox")
    modo.place(x=600, y=200)

    # Função para associar dropdowns
    def modo_selecionado(event):
        modo_selecionado = modo.get()
        if tipo_arquivo.get():
            selecionar_arq(modo_selecionado, root)

    # Associe o evento de mudança de seleção com a função
    modo.bind("<<ComboboxSelected>>", modo_selecionado)

    # Botões Avançar e Sair com tkinter Button
    btn_avancar = tk.Button(root, text="Avançar", font=("Arial", 14), bg="#007acc", fg="#ffffff", padx=30, pady=20, command=lambda: selecionar_arq(modo.get(), root))
    btn_avancar.place(x=100, y=700)

    btn_sair = tk.Button(root, text="Sair", command=root.destroy, font=("Arial", 14), bg="#007acc", fg="#ffffff", padx=30, pady=20)
    btn_sair.place(x=980, y=700)

    root.mainloop()
```

```python
if __name__ == "__main__":
    main()
```
